```python
file_path = 'sales Data.xlsx'
xls = pd.ExcelFile(file_path)
```

```python

df_list = []
for sheet_name in xls.sheet_names:
    df = xls.parse(sheet_name)
    df['Source Sheet'] = sheet_name  # Keep track of origin
    df_list.append(df)

sales_df = pd.concat(df_list, ignore_index=True)
```

```python
print("Initial Data Sample:\n", sales_df.head())
```

```python
sales_df.columns = sales_df.columns.str.strip().str.lower()
print("Cleaned columns:", sales_df.columns.tolist())
sales_df['order date'] = pd.to_datetime(sales_df['order date'], errors='coerce')
sales_df.dropna(subset=['order date'], inplace=True)
```

```python
total_revenue = sales_df['sales'].sum()
print(f"\nTotal Revenue: ${total_revenue:,.2f}")
product_sales = sales_df.groupby('product name')['sales'].sum().sort_values(ascending=False)
print("\nTop Selling Products:\n", product_sales)
sales_df['month'] = sales_df['order date'].dt.to_period('M')
monthly_sales = sales_df.groupby('month')['sales'].sum()
print("\nMonthly Sales:\n", monthly_sales)
```

```python
top_customers = sales_df.groupby('customer name')['sales'].sum().sort_values(ascending=False).head(10)

print("\nTop customers:\n", top_customers)
```

```python
product_sales.head(10).plot(kind='bar', title='Top 10 Products by Revenue')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python

monthly_sales.plot(kind='line', marker='o', title='Monthly Revenue Trend')
plt.ylabel('Revenue')
plt.xlabel('Month')
plt.tight_layout()
plt.show()
```

```python

top_customers.plot(kind='barh', title='Top 10 Customers by Revenue')
plt.xlabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
print("Numeric columns:", sales_df.select_dtypes(include=[np.number]).columns.tolist())
corr_matrix = sales_df.select_dtypes(include=[np.number]).corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```
